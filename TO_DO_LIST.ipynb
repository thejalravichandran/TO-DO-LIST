```python
from IPython.display import display
import ipywidgets as widgets
class ToDoList:
    def __init__(self):
        self.tasks = []
        self.task_list = widgets.VBox()
        self.input_box = widgets.Text(placeholder='Add a new task')
        self.add_button = widgets.Button(description='Add Task')
        self.add_button.on_click(self.add_task)

    def add_task(self, b):
        task_text = self.input_box.value
        if task_text:
            task_html = widgets.HTML(value=self.create_task_html(task_text))
            delete_button = widgets.Button(description='Delete', button_style='danger')
            delete_button.on_click(lambda x: self.delete_task(task_html, delete_button))

            task = widgets.HBox([task_html, delete_button])
            self.tasks.append(task)
            self.task_list.children = self.task_list.children + (task,)
            self.input_box.value = ''

    def create_task_html(self, task_text):
        return f'''
        <div style="display: flex; align-items: center;">
            <span style="cursor: pointer;" onclick="this.style.textDecoration = this.style.textDecoration === 'line-through' ? '' : 'line-through';">
                {task_text}
            </span>
        </div>
        '''

    def delete_task(self, task_html, delete_button):
        for task in self.tasks:
            if task_html in task.children:
                self.tasks.remove(task)
                break
        self.task_list.children = tuple(t for t in self.task_list.children if task_html not in t.children)

    def display(self):
        display(self.input_box)
        display(self.add_button)
        display(self.task_list)

# Create an instance of the ToDoList class and display it
todo = ToDoList()
todo.display()


```
